```python
import numpy as np
from openai import OpenAI

client = OpenAI()
```

```python
completion = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
    {"role": "system", "content": "You give short and best practice python code."},
    {"role": "user", "content": "Write a python function that uses regex to extract emails (including instances of 'name at email dot com') from a text. Use pytest to write useful unit tests."}
  ],
  temperature=0,
)

print(completion.choices[0].message.content)
```
